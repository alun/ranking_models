Formulas and tests for flat index for the upper right triangle.

```python
import pandas as pd


def flat(i, j, n):
    return int((i * (n - 1 + n - 1 - (i - 1)) / 2 + (j - i)) - 1)


def unflat(k, n):
    i = 0
    j = n - 1
    while k >= j:
        k -= j
        i += 1
        j -= 1
    return (i, i + 1 + k)


N = 10
pd.DataFrame([[flat(i, j, N) for j in range(N)] for i in range(N)])
```

```python
import numpy as np 

N = 100
flatten = pd.DataFrame([[flat(i, j, N) for j in range(N)] for i in range(N)])

comps = []
for k in range(int(N * (N - 1) / 2)):
    index = unflat(k, N)
    comps.append(k == flatten.to_numpy()[index])

np.all(comps)
```
